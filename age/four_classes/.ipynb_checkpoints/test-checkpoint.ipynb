```python
import cv2
from PIL import Image
import matplotlib.pyplot as plt
```

```python

location = [{'location': {'x1': 183, 'y1': 82, 'x2': 323, 'y2': 293}},\
            {'location': {'x1': 356, 'y1': 200, 'x2': 516, 'y2': 360}}, \
            {'location': {'x1': 333, 'y1': 143, 'x2': 420, 'y2': 332}}]
```

```python

image = cv2.imread("test.jpg")
image_copy = cv2.cvtColor(image.copy(), cv2.COLOR_BGR2RGB)


for index, i in enumerate(location):
    x1, x2, y1, y2 = i["location"]["x1"], i["location"]["x2"], i["location"]["y1"], i["location"]["y2"]
    
    cv2.imwrite(f"./input_images/{index}.jpg", image[y1:y2, x1:x2])
    cv2.rectangle(image_copy, (x1,y1), (x2, y2), (0,255,0), 2)

plt.figure(figsize=(16, 6.5))
plt.axis("off")
plt.imshow(image_copy)
plt.show()
```
